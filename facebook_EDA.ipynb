<h1>Data Cleaning(Preparation)</h1>

```python
fb_df.isnull()
```

```python
missing = fb_df.isnull().sum()
missing
```

```python
missing_percentage = fb_df.isnull().sum()/len(fb_df)*100
missing_percentage
```

```python
together = pd.concat([missing, missing_percentage], axis=1, keys=['missing','missing%'])
together
```

```python
fb_df = fb_df.replace({
        'gender': 0,
        'tenure': 0,
    }, np.nan)
fb_df
```

```python
sns.heatmap(fb_df.isnull())
```

```python
fb_df.isnull().sum()
```

```python
fb_df.shape
```

```python
fb_df_copy = fb_df.copy()
fb_df_copy.head()
```

```python
fb_df_copy = fb_df_copy.dropna(how='any', subset=['gender','tenure'])
fb_df_copy.head()
```

```python
fb_df_copy.shape
```

```python
fb_df_copy.isnull().sum()
```

```python
fb_df.duplicated().sum()
```

```python
fb_df_copy.shape
```

```python
fb_df_copy.to_csv('Pictures/fb_clean_df.csv')
```

<h1>Analysis and Visualization</h1>

```python
genders = fb_df_copy['gender'].value_counts()
genders
```

```python
plt.figure(figsize=(10,6))
plt.pie(genders, labels=['Male', 'Female'], autopct='%.2f%%', colors=['teal','cyan'])
plt.title('Males and Females')
plt.legend()
plt.show()
```

```python
ages = fb_df_copy['age'].value_counts()
ages
```

```python
sns.histplot(data = fb_df_copy, x='age', bins=30, kde=True)
plt.title('Age Distribution of Facebook Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
no_of_friends = fb_df_copy.groupby('gender')['friend_count'].sum().reset_index()
no_of_friends
```

<b>The number of friend of males and females are almost the same</b>

```python
no_of_likes_received = fb_df_copy.groupby('gender')['likes_received'].sum()
no_of_likes_received
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x=no_of_likes_received.index, y=no_of_likes_received.values)
plt.title('Number of likes received by Gender')
plt.xlabel('Gender')
plt.ylabel('No of likes received')
plt.show()
```

```python
no_of_likes = fb_df_copy.groupby('gender')['likes'].sum()
no_of_likes
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x=no_of_likes.index, y=no_of_likes.values)
plt.title('Number of likes by Gender')
plt.xlabel('Gender')
plt.ylabel('No of likes')
plt.show()
```

```python
age_grp=['10-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','101-110','111-120']
fb_df_copy['age_group']=pd.cut(fb_df_copy.age,bins=np.arange(10,121,10),labels=age_grp,right=True)
fb_df_copy.head()
```

```python
age_gender = fb_df_copy[['age_group','gender']].value_counts().reset_index()
age_gender
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=age_gender, x='age_group', y='count', hue='gender')
plt.title('Number of males and females in each group')
plt.xlabel('Age Groups')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=fb_df_copy, x='age_group', y='likes_received', hue='gender')
plt.title('Number of likes received in each group')
plt.xlabel('Age Group')
plt.ylabel('Likes Received')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=fb_df_copy, x='age_group', y='likes', hue='gender')
plt.title('Number of likes in each group')
plt.xlabel('Age Group')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=fb_df_copy, x='age_group', y='friend_count', hue='gender')
plt.title('Number of friends in each group')
plt.xlabel('Age Group')
plt.ylabel('Friend Count')
plt.show()
```
